```python
import time
from datetime import datetime, timedelta
def str_time_prop(start, end, format, prop):
    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))


def random_dates(start, end, prop, n):
    dates = []
    date = pd.to_datetime(str_time_prop(start, end, '%m/%d/%Y %I:%M %p', prop))
#     print(date)
    for i in range(0,n):
        dates.append((date + timedelta(days=7*i)).strftime("%Y-%m-%d"))
    return dates

def random_date(start, end, prop):
    d = pd.to_datetime(str_time_prop(start, end, '%m/%d/%Y %I:%M %p', prop))
    return d.strftime("%Y-%m-%d")
    
    

r = random_date("1/1/2008 1:30 PM", "1/1/2009 4:50 AM", random.random())
print(r)
# print(pd.to_datetime(r))
# print(r.strftime("%m/%d/%Y"))
```

```python
import pandas as pd
import random
import datetime
```

```python
no_entries = 425
no_audiofiles = 450
```

```python
df_users = pd.read_csv('./generated/users.csv')
df_users.head()
df_users['email']
```

```python
import time
from datetime import datetime, timedelta
def str_time_prop(start, end, format, prop):
    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))

def random_date(start, end, prop):
    d = pd.to_datetime(str_time_prop(start, end, '%m/%d/%Y %H:%M:%S', prop))
#     print(d)
    return d
```

```python
avg = int(no_entries/len(df_users['username']))
avg
```

```python
start_times = []
users = []
af_ids = []
total = 0
for user in df_users['email']:
    num_songs = random.randint(0,(avg*2-2))
    num_pods = random.randint(0,2)
    songs = []
    pods = []
    l = []
    for s in range(num_songs):
        songs.append(random.randint(1,300))
    for p in range(num_pods):
        pods.append(random.randint(301,no_audiofiles))
    total += num_songs + num_pods
    if len(songs) > 0:
        l =  [random.choice(songs) for _ in range(num_songs+1)] + pods
    for i in l:
#         print(i, user, random_date("1/1/2012 01:30:00", "1/1/2020 01:30:00", random.random()))
        start_times.append(random_date("1/1/2012 01:30:00", "1/1/2020 01:30:00", random.random()))
        users.append(user)
        af_ids.append(i)
        
#     print(user)
#     print(l)
print(total)
```

```python
df_streams = pd.DataFrame()
```

```python
df_streams['start_time']=start_times
df_streams['email']=users
df_streams['audiofile_id']=af_ids
df_streams
```
